# Import Libraries

```python
# Install the library
# %pip install -q great-expectations

# Import Library
from great_expectations.data_context import FileDataContext
```

# Instantiate Data Context

```python
# Create Data Context
context = FileDataContext.create(project_root_dir='./')
```

# Connect to a Datasource

```python
# Naming Datasource (Note: Need to be unique each datasources)
datasource_name = 'data-spotify-2010s'
datasource = context.sources.add_pandas(datasource_name)

# Naming Data Asset
asset_name = 'spotify-songs'
path_to_data = '/Users/ita/github-classroom/FTDS-assignment-bay/p2-ftds024-rmt-m3-dwitawin/P2M3_dwita_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build Batch Request
batch_request = asset.build_batch_request()
```

# Create Expectation Suite

```python
# Create Expectation Suite
expectation_suite_name = 'expectation-spotify-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a Validator based on Expectation Suite above
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check Validator
validator.head()
```

# Expectations

```python
# Expectation 1 : Column 'track_id' to be unique
validator.expect_column_values_to_be_unique('track_id')
```

```python
# Expectation 2 : Column 'track_name' to be unique
validator.expect_column_values_to_be_unique('track_name')
```

```python
# Expectation 3 : Column 'track_artist' can not be empty
validator.expect_column_values_to_not_be_null('track_artist')
```

```python
# Expectation 4 : Column 'track_popularity' must be between 0 and 100
validator.expect_column_values_to_be_between(
    column='track_popularity', min_value=0, max_value=100
)
```

```python
# Expectation 5 : Column 'track_album_id' can not be empty
validator.expect_column_values_to_not_be_null('track_album_id')
```

```python
# Expectation 6 : Column 'track_album_name' to be unique
validator.expect_column_values_to_be_unique('track_album_name')
```

```python
# Expectation 7 : Column 'track_album_release_date' match string time format
validator.expect_column_values_to_match_strftime_format(
    column='track_album_release_date',
    strftime_format='%Yyyy-%mm-%dd'
    )
```

```python
# Expectation 8 : Column 'playlist_name' to match regex
# Regex = 'Dance', 'Pop', 'EDM', 'Hip Hop', 'Remix', 'DJ', 'Rock', 'Upbeat', 'Latin', 'Cardio'
validator.expect_column_values_to_match_regex_list(
    column ='playlist_name',
    regex_list = ['Dance', 'Pop', 'EDM', 'Hip Hop', 'Remix', 'DJ', 'Rock', 'Upbeat', 'Cardio']
)
```

```python
# Expectation 9 : Column 'playlist_id' to be unique
validator.expect_column_values_to_be_unique('playlist_id')
```

```python
# Expectation 10 : Column 'playlist_genre' must contain one of the following of these things:
# set = 'pop', 'rap', 'rock', 'latin', 'r&b', 'edm'
validator.expect_column_values_to_be_in_set('playlist_genre', ['pop', 'rap', 'rock', 'latin', 'r&b', 'edm'])
```

```python
# Expectation 11 : Column 'playlist_subgenre' must contain one of the following of these things:
# set = ['dance pop', 'hip hop', 'post-teen pop', 'electropop', 'indie poptimism', 'trap', 'southern hip hop', 'gangster rap',
#  'album rock', 'tropical', 'latin pop', 'hip pop', 'classic rock', 'big room', 'permanent wave', 'pop edm', 'hard rock', 'reggaeton',
#  'latin hip hop', 'urban contemporary', 'new jack swing', 'neo soul', 'electro house', 'progressive electro house']

validator.expect_column_values_to_be_in_set('playlist_subgenre', [
       'dance pop', 'hip hop', 'post-teen pop', 'electropop',
       'indie poptimism', 'trap', 'southern hip hop', 'gangster rap',
       'album rock', 'tropical', 'latin pop', 'hip pop', 'classic rock',
       'big room', 'permanent wave', 'pop edm', 'hard rock', 'reggaeton',
       'latin hip hop', 'urban contemporary', 'new jack swing',
       'neo soul', 'electro house', 'progressive electro house'
])
```

```python
# Expectation 12 : Column 'danceability' must in form of integer or float
validator.expect_column_values_to_be_in_type_list('danceability',['integer', 'float'])
```

```python
# Expectation 13 : Column 'energy' must have max value of 1
validator.expect_column_max_to_be_between('energy', 0, 1)
```

```python
# Expectation 14 : Column 'key' must in form of integer
validator.expect_column_values_to_be_of_type('key', 'int64')
```

```python
# Expectation 15 : Column 'loudness' must in form of float
validator.expect_column_values_to_be_of_type('loudness', 'float64')
```

```python
# Expectation 16 : Column 'mode' must be in contain ONLY 0 and 1
validator.expect_column_values_to_be_in_set('mode',[0,1])
```

```python
# Expectation 17 : Column 'speechiness' to be not null
validator.expect_column_values_to_not_be_null('speechiness')
```

```python
# Expectation 18 : Column 'acousticness' in between 0 and 0,5
validator.expect_column_values_to_be_between('acousticness', 0, 0.5)
```

```python
# Expectation 19 : Column 'instrumentalness' max value is in between 0 and 1
validator.expect_column_max_to_be_between('instrumentalness', 0, 1)
```

```python
# Expectation 20 : Column 'liveness' max value is in between 0 and 1
validator.expect_column_max_to_be_between('liveness', 0, 1)
```

```python
# Expectation 21 : Column 'valence' max value is in between 0 and 1
validator.expect_column_max_to_be_between('valence', 0, 1)
```

```python
# Expectation 22 : Column 'tempo' max value is in between 0 and 250
validator.expect_column_max_to_be_between('tempo', 0, 250)
```

```python
# Expectation 23 : Column 'duration_sec' max value is in between 0 and 300
validator.expect_column_max_to_be_between('duration_sec', 0, 300)
```

```python
# Expectation 24 : Column 'duration_min' value is in between 0 and 10
validator.expect_column_values_to_be_between('duration_min', 0, 300)
```

```python
# Expectation 25 : Values in column 'duration_sec' is bigger than column 'duration_min'
validator.expect_column_pair_values_a_to_be_greater_than_b('duration_sec','duration_min',or_equal=True)
```

```python
# Save into Expectation Suite
validator.save_expectation_suite(discard_failed_expectations=False)
```

# Checkpoint

```python
# Create Checkpoint
checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator
)
```

```python
# Run a Checkpoint
checkpoint_result = checkpoint_1.run()
```
